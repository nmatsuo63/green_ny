```python
import pandas as pd
```

```python
# 学習用データと評価用データを読み込む
train_org = pd.read_csv('../data/smfg_train.csv', index_col=0)
```

```python
geo_train = train_org.copy()[['zip_city', 'boroname']]
```

```python
# 'zip_city'と'boroname'の組み合わせを作成
geo_train['combined_address'] = geo_train['zip_city'] + ', ' + geo_train['boroname']

# ユニークな組み合わせを抽出
unique_addresses = geo_train[['zip_city', 'boroname']].drop_duplicates()
unique_addresses['combined_address'] = unique_addresses['zip_city'] + ', ' + unique_addresses['boroname']
```

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import tqdm

# ジオコーディングの実行
geolocator = Nominatim(user_agent="geoapiExercises")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=2, error_wait_seconds=10)

```

```python
unique_addresses['location'] = unique_addresses['combined_address'].apply(lambda addr: geocode(addr))
# unique_addresses['latitude'] = unique_addresses['location'].apply(lambda loc: loc.latitude if loc else None)
# unique_addresses['longitude'] = unique_addresses['location'].apply(lambda loc: loc.longitude if loc else None)
```
